```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from plot_helpers import plot_learning_curve

# load in the file
path = 'penguin_dataset.csv'
df = pd.read_csv(path)

# clean up the file
df = df.drop(['individual_id','date_egg','sex', 'clutch_completion'],axis=1)

le = LabelEncoder()

df['island'] = le.fit_transform(df['island'])
df['species'] = le.fit_transform(df['species'])

df = df.dropna(axis=0)
print(df)

# split into train/test
train_x, test_x, train_y, test_y = train_test_split(df.drop(['species'],axis=1),df['species'],random_state=42,train_size=0.8,test_size=0.2,shuffle=True)

# normalize
scaler = StandardScaler()

scaler.fit(train_x)

train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)
```

```python
n = 1
while n <= 10:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(train_x, train_y)
    plt = plot_learning_curve(knn, f'{n} Neighbors',train_x,train_y)
    pred = knn.predict(test_x)
    type = 'micro'
    accuracy = accuracy_score(test_y,pred)
    precision = precision_score(test_y,pred,average=type)
    recall = recall_score(test_y,pred,average=type)
    print(f'{n} Neighbors --> Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}')
    n += 1
```

```python
n = 1
while n <= 10:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(train_x, train_y)
    plt = plot_learning_curve(knn, f'{n} Neighbors',train_x,train_y)
    pred = knn.predict(test_x)
    type = 'macro'
    accuracy = accuracy_score(test_y,pred)
    precision = precision_score(test_y,pred,average=type)
    recall = recall_score(test_y,pred,average=type)
    print(f'{n} Neighbors --> Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}')
    n += 2
```
